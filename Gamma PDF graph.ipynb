```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(start=0,stop = 100)
```

```python
y1 = np.exp(-x) #(alpha = 1,beta = 1)
y2 = (1/10)*np.exp(-x/10) #(alpha = 1, beta = 10)
y3 = (1/(24*5**5))*x**4*np.exp(-x/5) #(alpha = 5, beta = 5)
y4 = (1/362880)*x**9*np.exp(-x) #(alpha = 10, beta = 1)
```

```python
plt.plot(x,y1,x,y2,x,y3,x,y4)
plt.ylim((0,0.2))
plt.legend(labels = ("a = 1,b = 1","a = 1, b = 10","a = 5, b = 5","a = 10, b = 1"))
```
